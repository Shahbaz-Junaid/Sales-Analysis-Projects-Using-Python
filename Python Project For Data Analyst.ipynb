```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

# Add value labels on top of the bars
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Count of Customers by Age Group and Gender")
plt.tight_layout()
plt.show()
```

```python
sales_by_Age_group = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = "Age Group",y ="Amount",data = sales_by_Age_group)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the data
sales_by_Age_group = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Custom color list (make sure it matches the number of age groups)
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C299FF', '#FFD700', '#FF69B4']

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x="Age Group", y="Amount", hue="Age Group", data=sales_by_Age_group, palette=colors, legend=False)

plt.title("Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Sales Amount")
plt.tight_layout()
plt.show()
```

```python
# From above graphs we can see that most of the buyers are of age group between 26-35 yrs females and mens
```
